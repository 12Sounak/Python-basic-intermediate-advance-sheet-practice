```python
# 22.Write a Python Program to Find LCM?
def lcm(n,m):
    x=set()
    y=set()
    for i in range(1,n+1):
        if n % i == 0:
            x.add(i)
    for j in range(1,m+1):
        if m % j == 0:
            y.add(j)
    

    z= x.intersection(y) 




    p = max(z)
    return (n*m) // p
```

```python
# 23.Write a Python Program to Find HCF?
def lcm(n,m):
    x=set()
    y=set()
    for i in range(1,n+1):
        if n % i == 0:
            x.add(i)
    for j in range(1,m+1):
        if m % j == 0:
            y.add(j)
    

    z= x.intersection(y) 




    p = max(z)
    return p
```

```python
# 24.Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
# 0b10
def convert_decimal(number):
    binary = bin(number)
    octal = oct(number)
    hexadecimal = hex(number)
    
    return binary, octal, hexadecimal

# Input a decimal number from the user
decimal_number = int(input("Enter a decimal number: "))

# Convert the decimal number to binary, octal, and hexadecimal
binary, octal, hexadecimal = convert_decimal(decimal_number)

# Print the results
print(f"Binary representation of {decimal_number} is {binary}")
print(f"Octal representation of {decimal_number} is {octal}")
print(f"Hexadecimal representation of {decimal_number} is {hexadecimal}")
```

```python
# 26.Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?
def calculator(a, b):
    # Input the operation as a string
    ch = input("Enter the operation (+, -, *, /): ")

    if ch == '+':
        result = a + b
        print("The sum is", result)
    elif ch == '-':
        result = a - b
        print("The difference is", result)
    elif ch == '*':
        result = a * b
        print("The product is", result)
    elif ch == '/':
        if b != 0:
            result = a / b
            print("The quotient is", result)
        else:
            print("Error! Division by zero.")
    else:
        print("Invalid operation! Please enter one of +, -, *, /.")

# Input numbers from the user
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

# Call the calculator function
calculator(a, b)
```
