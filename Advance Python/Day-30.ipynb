```python
# 2. boolean_or
# boolean_or([True, True, False, False]) ➞ True 
# # [True, True, False, True] => [True, False, False] => [True, False] => True 
# perfectly working code:
def boolean_or(l):
    if len(l)<=1:
        return l
    l1 = l[::-1]
    l2=[]
    for i in range(0,len(l1)-1):
        l2.append(l[i] or l[i+1])
    return boolean_or(l2)
```

```python
# boolean_xor([True, True, False, False]) ➞ False 
# # [True, True, False, False] => [False, False, False] => [False, False] => False

# prefectly working code:
def boolean_xor(l):
    if len(l)<=1:
        return l
    l1 = l[::-1]
    l2=[]
    for i in range(0,len(l1)-1):
        l2.append(l[i] or l[i+1])
    return boolean_xor(l2)
```

```python
""" 
48. Create a function that creates a box based on dimension n. 
Examples 
make_box(5) ➞ [ 
  "#####", 
  "#   #", 
  "#   #", 
  "#   #", 
  "#####" 
] 
 
make_box(3) ➞ [ 
  "###", 
  "# #", 
  "###" 
] 
 
make_box(2) ➞ [ 
  "##", 
  "##" 
] 
 
make_box(1) ➞ [ 
  "#" 
]    
""" 
def make_box(n):
    print("#"*n)
    for i in range(n-2):
        print("#"+(" ")*(n-2)+"#")
    print("#"*n)
```
